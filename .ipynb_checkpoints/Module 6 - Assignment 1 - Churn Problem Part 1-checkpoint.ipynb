## Create a pipeline to impute empty strings and scale numircal values
In order to standardize and centralize the numerical values, a pipeline for imputing and scaling will be created. This will be done for the two numerical columns: TotalCharges and MonthlyCharges. After that, the distribution of the features will be shown using histograms.

```python
# Check for empty strings and transform them using imputer

# Print for each column the amount of empty strings for string columns
cols = list(df.select_dtypes(['object']).columns)

# Loop through string columns and replace empty string with nan
for x in cols:
    cond = df[x].str.contains(r'^\s*$', na=False)
    data = df.loc[cond, x]
    print(f"Number of rows containing whitespace for {x}: {sum(cond)}")
    if sum(cond) > 0:
        df[x] = df[x].replace(r'^\s*$', np.nan, regex = True)
```

```python
# Create pipeline for imputing and scaling
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
scaler = StandardScaler()
steps = [('imputer', imputer), ('scaler', scaler)]
pipeline = Pipeline(steps)

cols = ['MonthlyCharges', 'TotalCharges']

df[cols] = pipeline.fit_transform(df[cols])

df.head()

#print(f"Mean of TotalCharges non-scaled: {df.TotalCharges.mean()}")
#print(f"Standard deviation of TotalCharges non-scaled: {df.TotalCharges.std()}")
```

```python
# Create two histograms to show the columns have been scaled
fig, ax = plt.subplots(figsize = (8,6))

ax.hist(df.MonthlyCharges, color = 'red', alpha = 0.5, label = "Monthly charges")
ax.hist(df.TotalCharges, color = 'blue', alpha = 0.5, label = "Total charges")
ax.set(xlabel = "Monthly and total charges", ylabel = "Count")
ax.set_title("Distribution of scaled numerical values")
plt.legend()
plt.show()
```

__NOTE__: Although the boxplots basically look the same, the data is better represented for monthly and total charges

# Q1: Explain the process that needs to happen for each feature before you train your model.
Based on what we've seen so far on the data, is that a lot of features are covered as a Yes/No option or Categorical option. This has to be configured to numerical, since an algorithm can only deal with numerical data. In order to achieve this, the following two things need to happen:
- Convert features with two options (Yes/No, Male/Female etc.) to 1/0
- Convert features with more than two options to dummy variables. Drop one of the generated columns to avoid duplicate data

## Convert duo-optional features to 1/0

```python
# Select the columns with only 2 unique values
# Select the columns to check
cols_yes_no = [x for x in df.columns if len(df[x].unique()) == 2 and "Yes" in list(df[x].unique())]
cols_gender = ['gender']
mapping_yes_no = {'Yes': 1, 'No': 0}
mapping_gender = {'Female': 1, 'Male': 0}

# Remap gender and Yes/No columns
df[cols_yes_no] = df[cols_yes_no].apply(lambda x: x.replace(mapping_yes_no))
df[cols_gender] = df[cols_gender].apply(lambda x: x.replace(mapping_gender))
```

## Convert features with more than two options to dummy variables.

```python
# Select columns to use as dummy variables
colsDel = ['MonthlyCharges',  'TotalCharges', "customerID", "tenure"]
cols = list(df.columns)
cols = [x for x in cols if x not in colsDel and len(df[x].unique()) > 2]

# Create dummy variables with drop_first=True: df_clean
df_clean = pd.get_dummies(df, drop_first = True, columns = cols)

# Print shape to see new structure of the data
df_clean.shape
```

# Q2: Choose one metric to evaluate the different models you will train and explain why you are choosing that instead of other metrics. 
The target of this data is the Churn column: customers who left within the last month. This puts them into two categories. In order to predict the customers for in the future, you need to use a supervised classifier method. Based on the outcome of the predictions, the company may offer discounts to customers that might leave in order to retain them. Based on that, you want minimize the false negative: customers who leave but was predicted as not leave. In order to achieve that, the best metric to evaluate is __Recall: TP / (TP + FN)__
