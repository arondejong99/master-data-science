```python
# Import the modules
import sqlite3
import pandas as pd
import numpy as np
```

```python
path = "chinook.db"

# Setting up the connection using a function
def read_query(query, index_col = None):
    # Using context manager to automatically close connection
    with sqlite3.connect(path) as db:
        df = pd.read_sql_query(query, db, index_col = index_col)
    return df
```

## Query 1

```python
# How many customers are there that are not residing in the US?
query = "SELECT COUNT(*) AS num_customer_not_US FROM customers WHERE Country != 'USA'"
read_query(query)
```

## Query 3
Show Invoices of customers who are from the US. The resultant table should show the customer's full name, Invoice ID, Date of the invoice and billing country.

```python
query = """
    SELECT cus.FirstName, cus.LastName, inv.InvoiceId, inv.InvoiceDate, inv.BillingCountry
    FROM customers as cus
    INNER JOIN invoices as inv ON inv.CustomerId == cus.CustomerId
    WHERE cus.Country == 'USA'
"""
read_query(query)
```

## Query 7

## Query 8
How many tracks are there in each playlist?

```python
query = """
        SELECT pl.Name, pl_track.PlaylistId, COUNT(TrackId) AS num_track
        FROM playlist_track AS pl_track
        GROUP BY pl_track.PlaylistId
        INNER JOIN playlists AS pl ON pl.PlaylistId = pl_track.PlaylistId
        ORDER BY num_track
        """
read_query(query)
```

## Query 10
What is the sales agent that made the most sales in 2012?

```python
query = """
        SELECT emp.Title, emp.EmployeeId, emp.FirstName, emp.Lastname, SUM(inv.Total) AS total_sales 
        FROM employees AS emp
        INNER JOIN customers AS cus ON emp.EmployeeId == cus.SupportRepId
        INNER JOIN invoices AS inv ON cus.CustomerId == inv.CustomerId
        WHERE emp.Title == 'Sales Support Agent' AND inv.InvoiceDate LIKE '2012%'
        GROUP BY emp.EmployeeId
        ORDER BY total_sales DESC LIMIT 1        
        """
read_query(query)
```

## Query 14
Show the most purchased track of 2010

```python
query = """
    SELECT items.TrackId, tr.Name, COUNT(tr.TrackId) AS total
    FROM invoice_items AS items
    INNER JOIN tracks AS tr ON tr.TrackId == items.TrackId
    INNER JOIN invoices AS inv ON inv.InvoiceId == items.InvoiceId
    WHERE inv.InvoiceDate LIKE '2010%'
    GROUP BY tr.Name
    ORDER BY total DESC, tr.Name LIMIT 9
"""
read_query(query)
```

## Query 15
Show the top 5 most purchased tracks ever

```python
query = """
    SELECT items.TrackId, tr.Name, COUNT(tr.TrackId) AS total
    FROM invoice_items AS items
    INNER JOIN tracks AS tr ON tr.TrackId == items.TrackId
    GROUP BY tr.Name
    ORDER BY total DESC, tr.Name LIMIT 5
"""

read_query(query)
```

## Query 16
What are the top 3 best selling artists?

```python
query = """
WITH 
items_sold AS (
    SELECT tr.Name, tr.AlbumId
    FROM invoice_items AS items
    INNER JOIN tracks AS tr ON tr.TrackId == items.TrackId
    GROUP BY tr.Name
    ),
    albums_sold AS (
        SELECT alb.AlbumId, alb.ArtistId
        FROM albums AS alb
        INNER JOIN items_sold USING(AlbumId)
    ),
    artists_success AS(
        SELECT Name, COUNT(*) AS total_records_sold
        FROM artists
        INNER JOIN albums_sold USING(ArtistId)
        GROUP BY Name 
        ORDER BY total_records_sold DESC LIMIT 3
    )
    SELECT * FROM artists_success    
"""

read_query(query)
```

## Query 17
What is the most purchased Media Type?

```python
query = """
    SELECT med_t.Name, COUNT(*) AS total
    FROM invoice_items AS items
    INNER JOIN Tracks AS tr on tr.TrackId == items.TrackId
    INNER JOIN media_types AS med_t on med_t.MediaTypeId == tr.MediaTypeId
    GROUP BY med_t.Name
    ORDER BY total DESC LIMIT 1
"""
read_query(query)
```

## Query 18
Show the number of tracks purchased in all invoices that contain more than one genre.

```python
query = """
        SELECT InvoiceId, COUNT(tr.TrackId) AS num_tracks, COUNT(GenreId) AS num_genres
        FROM invoice_items AS items 
        INNER JOIN tracks AS tr ON tr.TrackId == items.TrackId 
        GROUP BY InvoiceId 
        HAVING COUNT(DISTINCT(tr.GenreId)) > 1
        ORDER BY num_genres
        """

read_query(query)
```
