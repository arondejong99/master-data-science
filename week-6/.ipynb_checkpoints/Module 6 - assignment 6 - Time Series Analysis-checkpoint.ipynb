# Introduction
In this exercise, a Time Series Analysis (TSA) will be performed. This will be done using the Retail Data Analytics data from Kaggle: https://www.kaggle.com/manjeetsingh/retaildataset 

This contains three different csv files that will be merged together.

```python
# Import the necessary modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error
```

```python
# Read the datasets
features = pd.read_csv("Features data set.csv", parse_dates = ['Date'])
sales = pd.read_csv("sales data-set.csv", parse_dates = ['Date'])
stores = pd.read_csv("stores data-set.csv")

display(features.head())
display(sales.head())
display(stores.head())

print(sales.groupby('Store')['Dept'].unique())
```

## Merge the datasets
In this paragraph, the datasets will be merged into one.

NOTE: feature `IsHoliday` will be used as key to prevent unnecessary extra features.

```python
df = pd.merge(features, sales, on = ['Store', 'Date', 'IsHoliday'], suffixes = ('_features', '_sales'))
df = pd.merge(df, stores, on = 'Store')

df.head()
```

```python
print(df.Dept.unique())
print(df.Store.unique())
```

## Clean the data
In this paragraph, the data will be cleaned. This will be done by doing the following steps:
- Deal with null values
- Deal with outliers
- Dealing with unexpected values

### Deal with null values
Based on the info method before, we can see that the Markdown1 to Markdown5 columns have a lot of null values. We'll check the correlation matrix to see if they have any linear relations with other features.

```python
corr = df.corr()

fig, ax = plt.subplots(figsize = (10, 10))

mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
sns.heatmap(corr, annot = True, cbar = False, 
            cmap = 'Blues', ax = ax, mask = mask)
plt.show()
```

#### Findings
Based on the correlation matrix, we can see that there're only a few linear relationships the Markdown column have with the other features. And since that there're a lot of missing values, the column will be removed from the dataset.

```python
# Remove MarkDown columns
cols = ['MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5']

df = df.drop(cols, axis = 1)

df.info()
```

- Unemployment

- Size

```python
cond = df['Weekly_Sales'] < 0
print(f"The amount of negative weekly sales values: {sum(cond)}")
```

### Findings
Based on the boxplots, we can see that the Weekly Sales is not really well distributed. Also, some values have a negative weekly sales, around 1285. This is odd, since the weekly sales indicates the amount of stuff sold in a week. That's why the rows that have a negative value for weekly sales will be removed. In order to combat the distribution problem, we'll normalise the values of the numeric columns when we're going to train our model.

```python
# Remove rows with a negative weekly sales value
df = df[df['Weekly_Sales'] >= 0]

print(f"The amount of negative weekly sales values: {sum(df['Weekly_Sales'] < 0)}")
```

# Data exploration and visualizations
- Sales to check for seasonality
- Sales compared to temperature
- Sales compared to unemployment rate
- Sales compared to fuel price
- Average sales for holidays vs non-holidays

Check if target has seasonality and what kind

```python
# Create function to compare the sales with a particular column in the dataset.
def sales(data, x, xcolumn, digits, ylabel, title):
    fig, ax = plt.subplots(figsize = (8, 6))
    # Copy the data
    df_copy = data.copy()
    # Round the numbers of x and set it in a new column of data
    df_copy[xcolumn] = np.round(data[x], digits)
    grouped = df_copy.groupby(xcolumn)['Weekly_Sales'].mean()
    grouped.plot()
    plt.xlabel(x)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.show()
    
# Create function to plot the sales per given period
def salesperiod(data, period, ax, title):
    # Set index to the datetime column
    data = data.resample(period, on = 'Date').size().reset_index(name='Count')
    
    sns.lineplot(data = data, x = 'Date',  y = 'Count', ax = ax, label = "Sales")
    ax.legend()
    ax.set_title(title)
```

```python
fig, axes = plt.subplots(figsize=(10, 16), nrows = 3)
titles = ["Sales per month", "Sales per quarter", "Sales per half year"]
periods = ["M", "3M", "2W"]

for ax, title, period in zip(axes.flat, titles, periods):
    salesperiod(df, period, ax, title)

plt.show()
```

```python
sales(df, "Temperature", 'Temperature_rounded', 0, 'Weekly Sales', "Temperature vs Sales")
```

```python
sales(df, "Unemployment", 'Unemployment_rounded', 0, 'Weekly Sales', "Unemployment rate vs Sales")
```

```python
sales(df, "Fuel_Price", 'Fuel_Price_Rounded', 1, 'Weekly Sales', "Fuel price in gallons vs Sales")
```

# Model training and predicting
## Scale the data

```python
# Select columns to scale the data
cols = ['Temperature', 'Fuel_Price', 'CPI', 'Unemployment', 'Weekly_Sales', 'Size']

# Instantiate Scaler
scaler = StandardScaler()

df[cols] = scaler.fit_transform(df[cols])
df.head()
```

## Selecting store and department

```python
# Select the store with the highest size and select department 1
df_clean = df.loc[(df.Size == max(df.Size)) & (df.Dept == 1), :].reset_index(drop = True)

# Drop the columns Store, Type and Dept because they're constant variables
cols = ['Store', 'Type', 'Dept']
df_clean = df_clean.drop(cols, axis = 1)
df_clean.head()
```

## Model training

```python
# Check if the dates are between 2010-02-05 and 2012-11-01.
print(f"Minimum date: {min(df_clean.Date)}")
print(f"Maximum date: {max(df_clean.Date)}")
```

```python
# Define train and test set
train = df_clean.loc[df_clean.Date < "2012-07-10", :].sort_values('Date').set_index('Date')
test = df_clean.loc[df_clean.Date >= "2012-07-10", :].sort_values('Date').set_index('Date')

# Define the input
X_train = train.drop('Weekly_Sales', axis = 1)
y_train = train.loc[:, 'Weekly_Sales']
X_test = test.drop('Weekly_Sales', axis = 1)
y_test = test.loc[:, 'Weekly_Sales']

y_train.head()
```

```python
# Instantiate and fit the model
model = SARIMAX(y_train, seasonal_order=(2,2,2,12))
model = model.fit()

# Predict on the test set
y_pred = model.get_forecast(len(test.index))
y_pred_df = y_pred.conf_int(alpha = 0.05) 

y_pred_df["Predictions"] = model.predict(start = y_pred_df.index[0], 
                                                end = y_pred_df.index[-1])
y_pred_df.index = test.index
y_pred_out = y_pred_df["Predictions"] 
```

```python
# Plot the train and test set
fig, ax = plt.subplots(figsize = (14, 10))

ax.plot(train.index, train['Weekly_Sales'], color = 'green', label = 'Train')
ax.plot(test.index, test['Weekly_Sales'], color = 'red', label = 'Test')
ax.plot(test.index, y_pred_out, color = 'blue', label = 'Prediction')
ax.set_xlabel("Date")
ax.set_ylabel("Weekly Sales")
ax.set_title("Sales per week")
ax.legend()
plt.show()
```

```python
# Calculate the RMSE to evaluate
rmse = np.sqrt(mean_squared_error(y_test, y_pred_out))
print(f"RMSE: {rmse:.5f}")
```

```python
# Instantiate and fit the model
model = SARIMAX(y_train, seasonal_order=(2,2,2,12))
model = model.fit()

# Predict on the test set
index = pd.date_range(test.index[0], "2013-03-30", freq = 'W')
y_pred = model.get_forecast(len(index))
y_pred_df = y_pred.conf_int(alpha = 0.05) 

y_pred_df["Predictions"] = model.predict(start = y_pred_df.index[0], 
                                             end = y_pred_df.index[-1])
y_pred_df.index = index
y_pred_out = y_pred_df["Predictions"] 
```

```python
# Plot the train and test set
fig, ax = plt.subplots(figsize = (14, 10))

ax.plot(train.index, train['Weekly_Sales'], color = 'green', label = 'Train')
ax.plot(test.index, test['Weekly_Sales'], color = 'red', label = 'Test')
ax.plot(y_pred_df.index, y_pred_out, color = 'blue', label = 'Prediction')
ax.set_xlabel("Date")
ax.set_ylabel("Weekly Sales")
ax.set_title("Sales per week")
ax.legend()
plt.show()
```
